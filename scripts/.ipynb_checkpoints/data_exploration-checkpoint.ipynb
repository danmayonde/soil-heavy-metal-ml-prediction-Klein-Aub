```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from essential_functions import descriptive_statistics, circular_descriptive_statistics, slope_aspect_direction
```

# Data Exploration

This notebook run statiscal analysis to provide an overview picture of our dataset.
In this notebook the following analysis are conducted on every features:

* Minimum and Maximum;
* Mean;
* Standard deviation;
* CV;
* Skewness;
* Kurtosis;
* Frequency counts for categorical features.

## Descriptive Statistics Analysis of the Soil Texture (Sand, Silt, and Clay)

```python
# Directory where the files should be saved
output_path = os.path.join("../Data", "Description_Stats")
```

```python
# Create a dataframe of soil texture data

soil_texture_df = pd.read_csv("../Data/Soil_Texture.csv")
soil_texture_df.head()
```

```python
# drop the unecessary features (Sample_ID, Longitude, Latitude) and conduct analysis
features_to_drop = ["Sample_ID", "Longitude", "Latitude"]

soil_texture_df = soil_texture_df.drop(features_to_drop, axis=1)

texture_stats, text_cat_stats = descriptive_statistics(soil_texture_df, output_filename="texture_stats.csv", save=True)

# show results
texture_stats
```

```python
# Log transformation

clay_df_np = soil_texture_df["Clay (%)"]
clay_df = pd.DataFrame(clay_df_np, columns=["Clay (%)"])
clay_df["claylog"] = np.log10(clay_df["Clay (%)"])

clay_stats,_ = descriptive_statistics(clay_df, output_filename="claylog.csv")

# Show results
clay_df.head()
clay_stats
```

```python

data = clay_df["Clay (%)"]  # variable to plot
x_label = "Clay (%)"
y_label = "Frequency"


data_clean = np.asarray(data).flatten()
data_clean = data_clean[~np.isnan(data_clean)]

# colors: steel blue fill, darker line
bar_color = "#4A90A4"
line_color = "#1B4965"
edge_color = "#2E6B7C"

fig, ax = plt.subplots(figsize=(6, 4), dpi=120)
n, bins, patches = ax.hist(
    data_clean,
    bins="auto",
    density=True,
    color=bar_color,
    edgecolor=edge_color,
    linewidth=0.6,
    alpha=0.85,
    label="Histogram",
)

# KDE line through the distribution
kde = stats.gaussian_kde(data_clean)
x_line = np.linspace(data_clean.min(), data_clean.max(), 200)
ax.plot(
    x_line,
    kde(x_line),
    color=line_color,
    linewidth=2.2,
    label="Density",
    zorder=5,
)

ax.set_xlabel(x_label, fontsize=12)
ax.set_ylabel(y_label, fontsize=12)
ax.legend(loc="upper right", frameon=True, fancybox=False, edgecolor="#e0e0e0", fontsize=10)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.tick_params(axis="both", labelsize=10)
plt.tight_layout()
plt.show()
```

```python
def slope_aspect_direction(df, output_filename='', save=False):
    """
    Add a 'Direction' column to a DataFrame based on angle values.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Input DataFrame with a single column containing angle values (in degrees)
    save : bool, default=False
        If True, save the DataFrame with the new Direction column to a CSV file
    output_path : str
        Path/filename for saving the CSV file (required if save=True)
    
    Returns:
    --------
    pandas.DataFrame
        DataFrame with the original column and a new 'Direction' column
    """
    # Make a copy to avoid modifying the original DataFrame
    aspect_direction_df = df.copy()
    
    # Get the single column name
    col_name = df.columns[0]
    
    # Create Direction column based on angle ranges
    def assign_direction(angle):
        if (angle >= 0 and angle <= 45) or (angle > 315 and angle < 360):
            return 'North'
        elif angle > 45 and angle <= 135:
            return 'East'
        elif angle > 135 and angle <= 225:
            return 'South'
        elif angle > 225 and angle <= 315:
            return 'West'
        else:
            return None  # For values outside 0-360 range
    
    aspect_direction_df['Direction'] = aspect_direction_df[col_name].apply(assign_direction)
    
    # Save to CSV if requested
    if save:
        output_dir = os.path.join('../Data', 'Description_Stats')
        output_path = os.path.join(output_dir, output_filename)
        aspect_direction_df.to_csv(output_path)
    
    return aspect_direction_df
```
